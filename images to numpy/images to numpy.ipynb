# Images to Numpy Array

```python
import os 
import numpy as np
from PIL import Image
```

```python
rootDir = 'photos'
image_opened = []
image_resized= []
img_np_array = []
```

```python
for dirName,subdirList,fileList in os.walk(rootDir):
    print('found directory . %s' %dirName)
    for fname in fileList:
        print('\t%s'% fname)
        img = Image.open(os.path.join(dirName,fname))
       # print(image_opened)
        
     #   print(img)
        res = img.resize((200,200))
   #     image_resized.append(res)
        img_np =np.array(res)
    #Stores the image into main Numpy Array
        img_np_array.append(img_np)
     #   print(img_np.shape)
```
